```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv(r"C:\Users\exam2\Downloads\text_classification.csv")
data
```

```python
#splitting the date
train,test = data[:],data[14:]
#extract vocabulary from training
vocabulary = []
for text in train['text']:
    text = text.lower()
    words = text.split()
    for word in words:
        if word not in vocabulary:
            vocabulary.append(word)
print(vocabulary)
```

```python
#make the table for both negative and positive
positive_set = train[train['status']=='pos']
negative_set = train[train['status']=='neg']
negative_set
```

```python
#term frequency matrix for positive documents
positive_data = {}
for text,index in zip(positive_set['text'],positive_set.index):
    positive_data[index] = []
    terms = text.lower().split()
    for word in vocabulary:
        positive_data[index].append(terms.count(word))
print(positive_set.index)
positive_table = pd.DataFrame(positive_data)
positive_table.index = vocabulary
positive_table = positive_table.T
positive_table
```

```python
#term frequency matrix for negative documents
negative_data = {}
for text,index in zip(negative_set['text'],negative_set.index):
    negative_data[index] = []
    terms = text.lower().split()
    for word in vocabulary:
        negative_data[index].append(terms.count(word))
print(negative_set.index)
negative_table = pd.DataFrame(negative_data)
negative_table.index = vocabulary
negative_table = negative_table.T
negative_table
```

```python
p,n,v = len(positive_set),len(negative_set),len(vocabulary)
p,n,v
```

```python
positive_prior = {}
negative_prior = {}
for word in vocabulary:
    positive_prior[word] = round((sum(positive_table[word].to_list())+1)/(p+v),2)
    negative_prior[word] = round((sum(negative_table[word].to_list())+1)/(n+v),2)
result = []
for text in test['text']:
    result.append(predict(text))
print(result)
```

```python
def predict(text):
    terms = text.lower().split()
    pos = 1
    neg = 1
    for term in terms:
        pos *= positive_prior[term]
        neg *= negative_prior[term]
    if pos>= neg:
        return 'pos'
    return 'neg'
```
